## 2. Is there a relationship between the price of a set and its number of pieces?

```python
relationship = df[["pieces","US_retailPrice"]]
relationship.head()
```

```python
relationship2 = relationship[relationship["US_retailPrice"].notna()]
relationship2
```

```python
relationship3 = relationship2[relationship2["pieces"].notna()]
relationship3
```

```python
korelasi = relationship3['US_retailPrice'].corr(df['pieces'], method='pearson')
korelasi
```

```python
korelasi2 = np.corrcoef(relationship3['US_retailPrice'], relationship3['pieces'])
```

```python
pearson_korelasi = korelasi2[0, 1]
pearson_korelasi
```

```python
import seaborn as sns
```

```python
plt.figure()
sns.scatterplot(x=relationship3['US_retailPrice'], y=relationship3['pieces'])
plt.title('Correlation between US Retail Price and Number of Pieces')
plt.xlabel('US Retail Price')
plt.ylabel('Number of Pieces')
plt.grid(True)
plt.show()
```

```python
# Setelah menghitung korelasi Pearson
if pearson_korelasi > 0.5:
    print("There is a strong positive relationship between the price of a set and the number of pieces.")
elif 0.3 < pearson_korelasi <= 0.5:
    print("There is a moderate positive relationship between the price of a set and the number of pieces..")
elif -0.3 <= pearson_korelasi <= 0.3:
    print("There is no linear relationship between the price of a set and the number of pieces.")
elif -0.5 <= pearson_korelasi < -0.3:
    print("There is a moderate negative relationship between the price of a set and the number of pieces.")
else:
    print("There is a strong negative relationship between the price of a set and the number of pieces.")
```

## 3. Which has been the most popular theme in each decade?

```python
dekade = df[["year","theme"]]
dekade
```

```python
def get_dekade(year):
    return str(year // 10 * 10) + "s"
```

```python
dekade['decade'] = dekade['year'].apply(get_dekade)
```

```python
dekade
```

```python
themes = dekade.groupby(['decade', 'theme']).size().reset_index(name='count')
```

```python
popular_theme = themes.loc[themes.groupby('decade')['count'].idxmax()]
```

```python
print(popular_theme)
```

```python
#data
themess = popular_theme['theme']
counts = popular_theme['count']
```

```python
plt.figure(figsize=(8, 6))
plt.bar(themess, counts, color= 'skyblue')
plt.title('Popular Theme in each decade')
plt.xlabel('themess')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```
